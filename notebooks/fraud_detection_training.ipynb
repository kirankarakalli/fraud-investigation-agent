```python
X=df.drop('Class',axis=1)
Y=df['Class']
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test=train_test_split(X,Y,test_size=0.2,stratify=Y)

```

```python
print(Y.value_counts(normalize=True))

print(y_train.value_counts(normalize=True))

print(y_test.value_counts(normalize=True))
```

```python
from sklearn.preprocessing import StandardScaler

scaler=StandardScaler()
X_train_scaled = X_train.copy()

X_train_scaled[['Amount', 'Time']] = scaler.fit_transform(
    X_train[['Amount', 'Time']]
)
```

```python
X_test_scaled = X_test.copy()

X_test_scaled[['Amount', 'Time']] = scaler.transform(
    X_test[['Amount', 'Time']]
)
```

```python
X_train_scaled.head()
```

```python
from sklearn.linear_model import LogisticRegression
lg=LogisticRegression()
```

```python
lg.fit(X_train_scaled,y_train)
```

```python
pred=lg.predict(X_test_scaled)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,pred))
```

```python
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier()

rf.fit(X_train_scaled,y_train)

```

```python
pred=rf.predict(X_test_scaled)
print(classification_report(pred,y_test))
```

```python
from sklearn.model_selection import RandomizedSearchCV

params = {
    'n_estimators': [50, 100],
    'max_depth': [10, 20, None],
    'min_samples_split': [2, 5],
    'min_samples_leaf': [1, 2],
    'max_features': ['sqrt']
}

random_search = RandomizedSearchCV(
    estimator=rf,
    param_distributions=params,
    n_iter=20,
    cv=3,
    scoring='recall', 
    random_state=42,
    n_jobs=-1
)

random_search.fit(X_train_scaled, y_train)
```

```python
random_pred=random_search.predict(X_test_scaled)
```

```python
print(classification_report(random_pred,y_test))
```

```python
print(random_search.best_params_)
print(random_search.best_score_)
best_model = random_search.best_estimator_
```
